```python
import pymc3 as pm 
import numpy as np
import matplotlib.pyplot as plt
```

plot up some distributions to see if they'd make good priors for our fits

```python
xs = np.linspace(0, 5.0)

ys = [np.exp(pm.Gamma.dist(alpha=3, beta=1.5).logp(x).eval()) for x in xs]
```

```python
plt.plot(xs, ys)
```
